```python
## importing library 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/Lecture file/25 th lecture 29 march/KNN Assignment/iphone_purchase_records.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.histplot(data['Purchase Iphone'], bins=20, kde=True, color='orange',)
plt.title("Distribution of Purchase Iphone by category")
plt.savefig('D:/Distribution of Purchase Iphone by category.png')
```

```python
sns.barplot(x='Purchase Iphone' ,y ='Age' ,hue ='Gender',data=data)
plt.title("purchase of iphone  based on age ")
plt.savefig('D:/purchase of iphone  based on age.png')
plt.show()
```

```python
sns.barplot(x='Purchase Iphone' ,y ='Salary' ,hue ='Gender',data=data)
plt.title(" purchase of iphone  based on salary")
plt.savefig('D:/ purchase of iphone  based on salary.png')
plt.show()
```

```python
d = {'Male':0 ,'Female':1}
data['Gender'] = data['Gender'].map(d)
```
